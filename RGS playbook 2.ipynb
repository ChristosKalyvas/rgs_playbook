## part 3.1 clean the data

```python
import pandas as pd

# Sample DataFrame
data = {
    'Company': ['Apple', 'Microsoft', 'Google', 'Amazon', 'Facebook'],
    'Ticker Symbol': ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'FB'],
    'Market Cap (Billion $)': [2256.98, 1779.26, 1685.66, 1602.91, 876.67],
    'Revenue (Billion $)': [274.52, 168.09, 182.53, 280.52, None],  # Example of missing value
    'Profit Margin': [0.21, 0.31, None, 0.05, 0.32]  # Example of missing value
}

df = pd.DataFrame(data)

# Print original DataFrame
print("Original DataFrame:")
print(df)

# Handling missing values
df.fillna({'Revenue (Billion $)': 0, 'Profit Margin': 0}, inplace=True)  # Fill missing values with 0

# Convert data types
df['Market Cap (Billion $)'] = df['Market Cap (Billion $)'].astype(float)
df['Revenue (Billion $)'] = df['Revenue (Billion $)'].astype(float)

# Remove whitespace from company names
df['Company'] = df['Company'].str.strip()

# Standardize ticker symbols to uppercase
df['Ticker Symbol'] = df['Ticker Symbol'].str.upper()

# Print cleaned DataFrame
print("\nCleaned DataFrame:")
print(df)
```

## Simple statistical approach standart deviation for outlier detection

```python
import numpy as np
from pyod.models.iforest import IForest

#the results dont seem to be correct

# a better approach would be to add the average of country
data_filled = df_emissions.fillna(0)
#print(data_filled['Scope 1 (metric tonnes CO2e)'])

X = data_filled['Scope 1 (metric tonnes CO2e)'].values.reshape(-1, 1)

# Train the Isolation Forest model
model = IForest(contamination=0.1)  # 10% expected outliers
model.fit(X)

# Predict outliers
outlier_scores = model.decision_function(X)
outliers = np.where(outlier_scores < 0)[0]

print("Indices of outliers:", outliers)
print("Values of outliers:", X[outliers])
```
