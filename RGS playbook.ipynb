## part 3.1 clean the data

```python
import pandas as pd

# Sample DataFrame
data = {
    'Company': ['Apple', 'Microsoft', 'Google', 'Amazon', 'Facebook'],
    'Ticker Symbol': ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'FB'],
    'Market Cap (Billion $)': [2256.98, 1779.26, 1685.66, 1602.91, 876.67],
    'Revenue (Billion $)': [274.52, 168.09, 182.53, 280.52, None],  # Example of missing value
    'Profit Margin': [0.21, 0.31, None, 0.05, 0.32]  # Example of missing value
}

df = pd.DataFrame(data)

# Print original DataFrame
print("Original DataFrame:")
print(df)

# Handling missing values
df.fillna({'Revenue (Billion $)': 0, 'Profit Margin': 0}, inplace=True)  # Fill missing values with 0

# Convert data types
df['Market Cap (Billion $)'] = df['Market Cap (Billion $)'].astype(float)
df['Revenue (Billion $)'] = df['Revenue (Billion $)'].astype(float)

# Remove whitespace from company names
df['Company'] = df['Company'].str.strip()

# Standardize ticker symbols to uppercase
df['Ticker Symbol'] = df['Ticker Symbol'].str.upper()

# Print cleaned DataFrame
print("\nCleaned DataFrame:")
print(df)
```

## part 3.2 merge the two datasets

```python
import pandas as pd

# Load market cap data
market_cap_df = pd.read_csv('market_cap_data.csv')  # Assuming you have a CSV file containing market cap data

# Load ESG metrics data
esg_metrics_df = pd.read_csv('esg_metrics_data.csv')  # Assuming you have a CSV file containing ESG metrics data

# Merge datasets on the 'symbol' column
merged_df = pd.merge(market_cap_df, esg_metrics_df, on='symbol', how='inner')

# Display the merged dataset
print(merged_df)

# Save the merged dataset to a CSV file
merged_df.to_csv('merged_market_cap_esg.csv', index=False)

print("Merged dataset created successfully.")
```
